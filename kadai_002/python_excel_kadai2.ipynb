```python
# （Colabなら多くは不要ですが保険で）
# !pip -q install pandas openpyxl

import pandas as pd

# 1) 元データ（文字列の日付でOK）
data = [
    {"日付": "2023-05-17", "社員名": "山田", "売上": 100, "部門": "メーカー"},
    {"日付": "2023-05-18", "社員名": "佐藤", "売上": 200, "部門": "代理店"},
    {"日付": "2023-05-19", "社員名": "鈴木", "売上": 150, "部門": "メーカー"},
    {"日付": "2023-05-20", "社員名": "田中", "売上": 300, "部門": "商社"},
    {"日付": "2023-05-21", "社員名": "高橋", "売上": 250, "部門": "代理店"},
]

df = pd.DataFrame(data)

# 2) 平均売上の算出（全体平均）
avg = df["売上"].mean()          # 例：200.0
avg_int = int(round(avg))        # 表示を揃えたいので整数に

# 「平均売上」列（全行に同じ平均値を入れる）
df["平均売上"] = avg_int

# 3) 業績ランク関数の定義
def performance(sales, avg_value):
    if sales >= avg_value + 50:
        return "A"
    elif sales >= avg_value:
        return "B"
    else:
        return "C"

# 「業績ランク」列の作成
df["業績ランク"] = df["売上"].apply(lambda x: performance(x, avg_int))

# 4) 列の並びを要件どおりに整える
df = df[["日付", "社員名", "売上", "部門", "平均売上", "業績ランク"]]

# 5) 新規Excelファイルに書き出し
out_path = "業績.xlsx"
with pd.ExcelWriter(out_path, engine="openpyxl") as writer:
    df.to_excel(writer, index=False, sheet_name="Sheet1")

print("書き出し完了:", out_path)
print(df)
```
